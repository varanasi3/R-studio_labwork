# TOPIC : using tidyverse and GGPLOT

```r
library(tidyverse)
```

```r
data<- (mpg) # loading data
head(data)
```

```r
ggplot(mpg,aes(x = displ,y = hwy))+geom_point() # ggplot with basic layer
```

```r
ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point()                   ## adding colors
```

```r
# Left column: x and y mapping needed!
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_point()

ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_smooth()

# Right column: no y mapping needed!
ggplot(data = mpg, aes(x = class)) +
  geom_bar()  

ggplot(data = mpg, aes(x = hwy)) +
  geom_histogram()
```

```r
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_point() +
  geom_smooth()
```

```r
ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point() +
  geom_smooth(se = FALSE)

# color aesthetic specified for only the geom_point layer
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_point(aes(color = class)) +
  geom_smooth(se = FALSE)
```

## bar graph 

```r
ggplot(mpg , aes(x=class))+geom_bar()
```

```r
class_count <- dplyr::count(mpg, class)
class_count
```

## stacked bar graph


```r
ggplot(mpg, aes(x = class, fill = drv)) + 
  geom_bar()
```

```r
ggplot(mpg, aes(x = class, fill = drv)) + 
  geom_bar(position = "dodge")

# position = "fill": percentage chart
ggplot(mpg, aes(x = class, fill = drv)) + 
  geom_bar(position = "fill")
```

## Line plot

```r
ggplot(mpg, aes(displ, hwy)) + 
  geom_point(color = "grey") + 
  stat_summary(fun.y = "mean", geom = "line", size = 1, linetype = "dashed")
```

```r
ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point() +
  scale_color_brewer()

# specifying color palette
ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point() +
  scale_color_brewer(palette = "Set3")
```

## Pie chart

```r
counts <- data.frame(table(mpg$class))  #  loading another dataset
counts
```

```r
ggplot(counts, aes(x="", y=Freq, fill=Var1)) +
  geom_bar(stat="identity", width=1) +
  coord_polar(theta="y") +
  scale_fill_brewer(palette="Set3") +
  theme_void() +
  labs(title="Pie Chart of Car Classes in mpg Dataset", fill="Class")
```

## Area plot

```r


ggplot(mpg, aes(x = hwy)) +
  geom_density(fill = "blue", alpha = 0.5) +
  labs(title = "Density of Highway MPG")
```

## Box plots

```r
ggplot(mpg, aes(x=class, y=hwy)) + 
  geom_boxplot() + 
  labs(title="Highway mileage by car class", x="Car class", y="Highway mileage")
```

## Violin plots

```r
ggplot(mpg, aes(x = factor(cyl), y = cty)) +
  geom_violin() +
  labs(x = "Number of Cylinders", y = "City MPG") +
  ggtitle("City MPG by Number of Cylinders") +
  theme_bw()
```

## scatter plot

```r
ggplot(mpg, aes(x = class, y = hwy)) +
  geom_point() +
  labs(title = "Highway Miles per Gallon by Vehicle Class",
       x = "Vehicle Class",
       y = "Highway MPG")
```
